```python
u_data= Unemployment[Unemployment["Area"] =="Urban"]
r_data= Unemployment[Unemployment["Area"] =="Rural"]
```

```python
u_data.reset_index(inplace=True,drop=True)
```

```python
r_data.reset_index(inplace=True,drop=True)
```

```python
def new_data(urban, rural):
    new_Data = pd.DataFrame()  # Initialize the DataFrame outside the loop

    # Get unique regions from both DataFrames
    regions = pd.concat([urban["Region"], rural["Region"]]).unique()

    for region in regions:
        # Select rows for the current region from urban and rural DataFrames
        urban_region = urban[urban["Region"] == region]
        rural_region = rural[rural["Region"] == region]

        # Concatenate rows for the current region into new_Data
        new_Data = pd.concat([new_Data, urban_region, rural_region], ignore_index=True)

    return new_Data
new_Data = new_data(u_data, r_data)
new_Data
```

```python
regions = pd.concat([u_data["Region"], r_data["Region"]]).unique()
len(regions)
```

```python
sns.pointplot(data=new_Data1,x=new_Data1[" Date"],hue=new_Data1["Region"],y=new_Data1[" Estimated Unemployment Rate (%)"]);
plt.xticks(rotation=90)
plt.legend(title="Region", loc='upper left')
plt.show()

sns.pointplot(data=new_Data2,x=new_Data2[" Date"],hue=new_Data2["Region"],y=new_Data2[" Estimated Unemployment Rate (%)"]);
plt.xticks(rotation=90)
plt.legend(title="Region", loc='upper left')
plt.show()

sns.pointplot(data=new_Data3,x=new_Data3[" Date"],hue=new_Data3["Region"],y=new_Data3[" Estimated Unemployment Rate (%)"]);
plt.xticks(rotation=90)
plt.legend(title="Region", loc='upper left')
plt.show()

sns.pointplot(data=new_Data4,x=new_Data4[" Date"],hue=new_Data4["Region"],y=new_Data4[" Estimated Unemployment Rate (%)"]);
plt.xticks(rotation=90)
plt.legend(title="Region", loc='upper left')
plt.show()

```
